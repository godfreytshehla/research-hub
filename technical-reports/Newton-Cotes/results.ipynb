```python
from newton_cotes import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

# $f(x) = e^{-x} \cdot \sin(10x)$

```python
def f1(x):
    return np.exp(-x) * np.sin(10 * x)
a, b = 0, 2
reference1, _ = quad(f1, a, b)
n_values = np.arange(2, 101, 2)
trapezoidal_errors1 = []
simpsons_errors1 = []
for n in n_values:
    trapezoidal_result = trapezoidal_rule(f1, a, b, n)
    simpsons_result = simpsons_rule(f1, a, b, n)
    trapezoidal_errors1.append(abs(trapezoidal_result - reference1))
    simpsons_errors1.append(abs(simpsons_result - reference1))
plt.figure(figsize=(10, 6))
plt.plot(n_values, trapezoidal_errors1, label="Trapezoidal Rule", marker="o")
plt.plot(n_values, simpsons_errors1, label="Simpson's Rule", marker="s")
plt.xlabel("Number of Subintervals (n)")
plt.ylabel("Absolute Error")
plt.title("Convergence for $f(x) = e^{-x} \cdot \sin(10x)$")
plt.legend()
plt.grid()
plt.savefig("oscillatory_exponential_decay.png", dpi=300, bbox_inches="tight")  # Save the plot
plt.show()
```

```python
def f2(x):
    return 1 / np.sqrt(1 + x**4)
reference2, _ = quad(f2, a, b)
trapezoidal_errors2 = []
simpsons_errors2 = []
for n in n_values:
    trapezoidal_result = trapezoidal_rule(f2, a, b, n)
    simpsons_result = simpsons_rule(f2, a, b, n)
    trapezoidal_errors2.append(abs(trapezoidal_result - reference2))
    simpsons_errors2.append(abs(simpsons_result - reference2))
plt.figure(figsize=(10, 6))
plt.plot(n_values, trapezoidal_errors2, label="Trapezoidal Rule", marker="o")
plt.plot(n_values, simpsons_errors2, label="Simpson's Rule", marker="s")
plt.xlabel("Number of Subintervals (n)")
plt.ylabel("Absolute Error")
plt.title("Convergence for $f(x) = \\frac{1}{\sqrt{1 + x^4}}$")
plt.legend()
plt.grid()
plt.savefig("rational_singularity.png", dpi=300, bbox_inches="tight")  # Save the plot
plt.show()
```
